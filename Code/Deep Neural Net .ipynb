# Using a deep nural networ to make a mediocre prediction

```python
import pandas as pd
import numpy as np

```

## We import and pre- process our data 

```python
cookies = pd.read_csv("cookies.csv")
```

```python
#we drop null values
cookies = cookies.dropna()
```

```python
#we drop variables we didn't feel were worth it based on our data exploration 

cookies = cookies.drop(columns= ["crunch factor", "density", "diameter", "pH", "aesthetic appeal"])
```

```python
#We create dummie variables to fix up the data 

one = pd.get_dummies(cookies.mixins)
one = one.drop(columns=["peanut butter, raisins", "oats"])
#We now join it into the new dataframe 
cookies = pd.concat([cookies, one], axis=1)
```

```python
# We did the same for butter type 
two = pd.get_dummies(cookies["butter type"])
cookies= pd.concat([cookies, two], axis=1)
```

```python
#We Dropped the two columns we have encoded.
cookies = cookies.drop(columns=["butter type", "mixins"])
```

## We find and drop the outliers that will cause unnecessary noise 

```python
#finding the outliers

from pyod.models.knn import KNN
import pandas as pd
X = pd.DataFrame(data=cookies)
clf = KNN(contamination=0.01)
clf.fit(X)
y_pred = clf.predict(X)
cookies_outliners = X[y_pred == 1]
cookies_outliners
```

```python
# dropping the outliers 
for x in cookies_outliners.index:
    cookies_not_outliners = cookies.drop([x],axis = 0,inplace = True)
```

## Splitting the data

```python
#We now split our data. the cometed parts of the text is simply where we were spliting our data to test. 

from sklearn import preprocessing

X= cookies.drop(columns="quality")
y=cookies["quality"]

from sklearn.model_selection import train_test_split

#X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state = 8)


min_max_scaler = preprocessing.MinMaxScaler()


## Scaling and transforming our data: 

X_scaled = min_max_scaler.fit_transform(X.values)
X_train = pd.DataFrame(X_scaled)
X_train

#X_scaled_t = min_max_scaler.transform(X_test.values)
#X_test = pd.DataFrame(X_scaled_t)
```

```python
# Scaling our y into values between 0 and 1 to accomodate the model. It works best with values between 0 and 1
y = y/11
y.value_counts()
```

```python
#y_test = y_test/11
#y_test.value_counts()
```

# Assembling our net

```python
#we import Keras 

from tensorflow import keras
from tensorflow.keras import layers

#We define our layers and the density of our model
#input shape is the number of coolumns of our data frame 

model = keras.Sequential([
    layers.Dense(512, activation='relu', input_shape=[21]),
    layers.Dense(512, activation='relu'),
    layers.Dense(512, activation='relu'),
    layers.Dense(1),
])

#We compile and tell the model what to optimize for 

model.compile(
    optimizer='adam',
    loss='mse',
)

# we fit our model to our X and y, history is so that we can see the model work. 
# epochs is the amout of times the model will repeat, we found 50 to be our sweet spot 

#We also stoped adding and removing layers since it was making our performace worse off

model_tracking = model.fit(
    X_train, y,
    #validation_data=(X_test, y_test),
    batch_size=500,
    epochs=50,
)

```

Keep in mind that for this we had split our original data since there was no way to track our progress otherwise
